```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

import seaborn as sns
```

```python
loop = 1000
qn = 300
filename = f"C:/Users/Admin/Desktop/ELO_Rating/Elo-Simulation/src/data/Actual Results {loop}loop {qn}qn.csv"
df = pd.read_csv(filename)
```

```python
df
```

```python
acc03 = df["0.3"]
acc03
```

```python
acc03_mean = np.mean(acc03)
acc_03_std = np.std(acc03)
pdf = stats.norm.pdf(acc03, acc03_mean, acc_03_std)
plt.plot(acc03, pdf)
plt.title("Elo Rating Distribution for C with accuracy = 0.3")
```

```python
def show_pdf_normal(item):
    person = df[item]
    person_mean = np.mean(person)
    person_std = np.std(person)
    pdf = stats.norm.pdf(person, person_mean, person_std) # seaborn --> pdf, actual distribution
    plt.plot(person, pdf)
    plt.axvline(x=person_mean, color='r', linestyle='--', linewidth=2)
    plt.text(person_mean + 15, 0.001, f"mu = {round(person_mean)}", color='r')
    plt.title(f"Elo Rating Distribution for C with accuracy = {item} (Nloop = {loop}, NQn = {qn})")
```

```python
def show_pdf_actual(item):
    sns.kdeplot(df[item], fill=True)
    person_mean = np.mean(df[item])
    plt.xlabel("Elo Rating")
    plt.ylabel("Density")
    plt.axvline(x=person_mean, color='r', linestyle='--', linewidth=1)
    plt.text(person_mean + 15, 0.001, f"mu = {round(person_mean)}", color='r')
    plt.title(f"Elo Rating Distribution for C with accuracy = {item} (Nloop = {loop}, NQn = {qn})")
```
